```python

import scrapy 
class QuotesSpider(scrapy.Spider): 
    name = "quotes" 
    start_urls = ["http://quotes.toscrape.com/"] 
 
    def parse(self, response): 
        for quote in response.css("div.quote"): 
            yield { 
                "text": quote.css("span.text::text").get(), 
                "author": quote.css("small.author::text").get() 
            } 
        next_page = response.css("li.next a::attr(href)").get() 
        if next_page: 
            yield response.follow(next_page, self.parse)
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Simulated HTML snippet (you'd fetch this via requests or Selenium)
html = requests.get("https://coinmarketcap.com/")

soup = BeautifulSoup(html.text, 'html.parser')

# Find all rows
rows = soup.find_all('tr', style='cursor:pointer')

data = []

for row in rows:
    # Extract coin name
    name_tag = row.find('p', class_=lambda x: x and 'coin-item-name' in x)
    coin_name = name_tag.text.strip() if name_tag else None

    # Extract price using $ symbol
    price_tag = row.find('td', style='text-align:end')
    price_span = price_tag.find('span') if price_tag else None
    price = price_span.text.strip() if price_span and '$' in price_span.text else None

    if coin_name and price:
        data.append({'Coin': coin_name, 'Price': price})

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```
